````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}

scores = pd.DataFrame(raw_data)
print(scores)
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(8, 6))
plt.scatter(scores['preTestScore'], scores['postTestScore'], s=scores['age']*2, alpha=0.5)

plt.title('Scatter Plot of preTestScore vs postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(8, 6))
plt.scatter(scores['preTestScore'], scores['postTestScore'], s=scores['postTestScore']*4.5, c=scores['female'], cmap='viridis', alpha=0.5)

plt.title('Scatter Plot of preTestScore vs postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

plt.colorbar(label='Female (0: Male, 1: Female)')

plt.show()
````

### BONUS: Create your own question and answer it.

How does the distributionof post-test scores differ between male and female participants?

````python
male_scores = scores[scores['female'] == 0]['postTestScore']
female_scores = scores[scores['female'] == 1]['postTestScore']


plt.figure(figsize=(10, 6))
plt.hist(male_scores, bins=10, alpha=0.5, label='Male')
plt.hist(female_scores, bins=10, alpha=0.5, label='Female')
plt.xlabel('Post-test Score')
plt.ylabel('Frequency')
plt.title('Distribution of Post-test Scores by Gender')
plt.legend()
plt.grid(True)
plt.show()
````
