## [1] datetime 오브젝트

datetime 라이브러리는 날짜와 시간을 처리하는 등의 다양한 기능을 제공하는 파이썬 라이브러리.
 - 날짜를 처리하는 date 오브젝트
 - 시간을 처리하는 time 오브젝트
 - 날짜와 시간을 모두 처리하는 datetime 오브젝트

```python
from datetime import datetime

#now,today 메서드를 사용하면 다음과 같이 현재 시간을 출력할 수 있음.
now1 = datetime.now()
print(now1)
```

```python
now2 = datetime.today()
print(now2)
```

```python
#datetime 오브젝트를 생성할 때 시간을 직접 입력하여 인자로 전달한 것. 
t1 = datetime.now()
t2 = datetime(1970,1,1)
t3 = datetime(1970,12,12,13,24,34)

print(t1)
print(t2)
print(t3)
```

```python
#datetime 오브젝트로는 시간계산을 할 수 있음. 
diff1 = t1 - t2
print(diff1)
print(type(diff1))
```

```python
diff2 = t2 - t1
print(diff2)
print(type(diff2))
```

**[dt 접근자 사용하기]**

 : datetime 오브젝트에서도 문자열을 처리하기 위해서 dt접근자를 사용함. 
 datetime 속성이나 메서드를 사용하여 시계열 데이터를 처리할 수 있음. 

```python
#Date열을 datetime 오브젝트로 변환하여 새로운 열로 추가.
ebola = pd.read_csv('https://raw.githubusercontent.com/mindcompass/data_sets/master/country_timeseries.csv')
ebola['date_dt'] = pd.to_datetime(ebola['Date'])

#dt접근자를 사용하지 않고 인덱스가 3인 데이터의 년, 월, 일 데이터를 추출
print(ebola[['Date','date_dt']].head())

print(ebola['date_dt'][3].year)
print(ebola['date_dt'][3].month)
print(ebola['date_dt'][3].day)
```

위와 같은 방법은 date_dt열의 특정 데이터를 인덱스로 접근해야 하기 때문에 불편함. 

```python
#dt 접근자로 date_dt열에 한번에 접근
ebola['year'] = ebola['date_dt'].dt.year
print(ebola[['Date','date_dt','year']].head())
```

```python
ebola['month'],ebola['day'] = (ebola['date_dt'].dt.month, ebola['date_dt'].dt.day)
print(ebola[['Date','date_dt','year','month','day']].head())
```

```python
#새로 추가한 열의 자료형 출력
print(ebola.info())
```

* 그래프를 그리기 위한 데이터프레임 준비하기

중간에 아예 날짜가 없는 데이터가 있어 이 데이터도 포함시켜야 확산 속도를 제대로 비교할 수 있음.

```python
ebola = pd.read_csv('https://raw.githubusercontent.com/mindcompass/data_sets/master/country_timeseries.csv', parse_dates=['Date'])

print(ebola.head().iloc[:,:5])
```

```python
print(ebola.tail().iloc[:,:5])
```

```python
#Date열을 인덱스를 지정하고, Date열의 최댓값과 최솟괍으로 시간범위를 생성하여 new_idx에 저장.
ebola.index = ebola['Date']
new_idx = pd.date_range(ebola.index.min(),ebola.index.max())
```

```python
#new_idx는 시간 순서와 반대로 생성되어 있어 reversed 메서드로 인덱스를 반대로 뒤집음.
print(new_idx)

new_idx = reversed(new_idx)
```

reindex메서드를 사용하여 새로 생성한 인덱스를 새로운 인덱스로 지정

```python
ebola = ebola.reindex(new_idx)

print(ebola.head().iloc[:,:5])
```

```python
print(ebola.tail().iloc[:,:5])
```

각 나라의 에볼라 발병일 옮기기

last_valid_index, first_valid_index 메서드를 사용하여 각 나라의 에볼라 발병일을 구함.

```python
last_valid = ebola.apply(pd.Series.last_valid_index)
print(last_valid)
```

```python
first_valid = ebola.apply(pd.Series.first_valid_index)
print(first_valid)
```

각 나라의 에볼라 발병일을 동일한 출발선으로 옮기려면 에볼라가 가장 처음 발병한 날에서 각 나라의 에볼라 발병일을 뺀 만큼만 옮기면 됨.

```python
earliest_date = ebola.index.min()
print(earliest_date)
```

```python
shift_values = last_valid - earliest_date
print(shift_values)
```

```python
ebola_dict = {}
for idx, col in enumerate(ebola):
    d = shift_values[idx].days
    shifted = ebola[col].shift(d)
    ebola_dict[col] = shifted
```

```python
ebola_shift = pd.DataFrame(ebola_dict)
```

```python
print(ebola_shift.tail())
```

```python
#인덱스를 Day 열로 지정하고 그래프에 필요없는 열은 삭제
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date','Day'],axis=1)

print(ebola_shift.tail())
```

```python
fig,ax = plt.subplots()
ax = ebola_shift.iloc[:,:].plot(ax=ax)
ax.legend(fontsize=7,loc=2,borderaxespad=0.)
plt.show()
```
