### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv", sep=",")
```

### Step 4. Delete the Unnamed 0 column

```python
if 'Unnamed: 0' in tips.columns:
    tips.drop('Unnamed: 0', axis=1, inplace=True)
    
print(tips.head())
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(tips['total_bill'], bins=20, color='skyblue', alpha=0.7)

plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(10, 6))
plt.scatter(tips['total_bill'], tips['tip'], color='skyblue', alpha=0.7)

plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(10, 6))
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*20, alpha=0.5)

plt.title('Relationship between Total Bill, Tip, and Size')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(['Size'])

plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(10, 6))
plt.boxplot([tips[tips['day'] == 'Thur']['total_bill'],
             tips[tips['day'] == 'Fri']['total_bill'],
             tips[tips['day'] == 'Sat']['total_bill'],
             tips[tips['day'] == 'Sun']['total_bill']],
            labels=['Thur', 'Fri', 'Sat', 'Sun'])

plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')

plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(10, 6))
colors = {'Male': 'blue', 'Female': 'orange'}
for sex, color in colors.items():
    plt.scatter(tips[tips['sex'] == sex]['tip'], tips[tips['sex'] == sex]['day'], color=color, alpha=0.7)

plt.title('Scatter Plot of Tip vs Day, by Sex')
plt.xlabel('Tip ($)')
plt.ylabel('Day of the Week')

plt.legend(colors.keys())

plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='Set2')

plt.title('Total Bill per Day, Differentiated by Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')

plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.set_style("whitegrid")

g = sns.FacetGrid(tips, col="time", height=6)
g.map(plt.hist, "tip", bins=15, color='skyblue', alpha=0.7)

g.set_axis_labels("Tip ($)", "Frequency")
g.set_titles("{col_name}")

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']

sns.set_style("whitegrid")

plt.figure(figsize=(15, 6))

# Scatterplot for Male
plt.subplot(1, 2, 1)
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=male_data, palette='Set1')
plt.title('Total Bill vs Tip (Male)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Smoker', loc='upper right')

# Scatterplot for Female
plt.subplot(1, 2, 2)
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=female_data, palette='Set1')
plt.title('Total Bill vs Tip (Female)')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Smoker', loc='upper right')

plt.show()
```
